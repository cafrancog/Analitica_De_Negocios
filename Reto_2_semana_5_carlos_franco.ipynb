1. Obtener información sobre las variables.

En las especirficaciones del trabajo se pide que se tomen 8 variables de caracterización de los empleados, excluyendo la variable de personalidad, para aumentar el entendimiento propio y del espectador, vamos a describir cada una y su significado aquí.



*   Business: esta se entiende como la linea de negocios en la que está ubicada esta persona ya sea hardware que se representará con un 0, o software que se representará con un 1.

*   Age: esta es la edad que reporta la persona y se muestra en números discretos (numeros enteros sin decimales)

* Female: esta indica si la persona es de género femenino, el número 1 indica que la persona es una mujer, mientras que el 0 indica el caso contrario, osea que la persona es un hombre.

* Years: Esta es la cantidad de años que una persona ha trabajado continuamente para la compañía e igual que la edad se trabaja de con números enteros discretos.

* College: Si esta persona se ha graduado de educación universitaria de cuatro años, se representa 1 como si, y 0 como que no.

* Certificates: Este es el número de certificados de competencias profesionales que esta persona posee, esta tambien se expresa en números enteros.

* Feedback: la nota que ha recibido este empleado por parte de sus compañeros y supervisor de forma anual, la menor puntuación posible es cero (0) y la mayor es cuatro (4), esta escala va de 1 a 4 siendo 4 representante de un mejor rendimiento, y 1 de un rendimiento cuestionable, esta variable es numérica discreta, osea que si se encuentran decimales.


* Salary: este es el salario anual que gana cada empleado

* NPS: Este es el Net promoter score o Puntuación neta del promotor, este mide la lealtad del cliente. Según Qualtrics (2024), el NPS pregunta al cliente si recomendaría el producto, empresa o servicio a conocidos, el resultado de la respuesta de todos los clientes varía de -100 que significa que el 100% no recomendaría, a 100 que significa que el 100% recomendaría.
fuente: https://www.qualtrics.com/experience-management/customer/measure-nps/

Sin embargo, en este caso solo se tienen datos de 1 a 10 por lo que se va a asumir que en vez de una escala a 100 se utiliza una a 10, y que la ausencia de números negativos significa que ninguna medida NPS ha sido negativa, no que se mida de una forma distinta.

3. Cargar la base de datos

```python


bdatos = pd.read_excel("/content/2. TechSalesReport_Int_L.xlsx", sheet_name = 0 )

print(bdatos.head())
```

4. Transformar datos

```python


bdatos["College"] = bdatos["College"].map({"Yes": 1, "No": 0})

bdatos['Business'] = bdatos["Business"].map({"Hardware": 0, "Software": 1})
print(bdatos.head())
```

5. Verificar integridad de los datos

```python
print(pd.isna(bdatos).sum()) #no hay ningun NAN, no se hace nada mas

college = len(bdatos["College"])
college_true = len(bdatos[bdatos["College"] == 1])

print("Total de datos", college)

print("Gente con educacion universitaria", college_true)

print("el porcentaje de gente en la base de datos con titulo universitario es:",((college_true / college)*100))
```

Se encuentra un gran sesgo en la base de datos ya que mucho mas del 50% tienen educación universitaria, pero ante la falta de conocimiento de si este modelo es afectado como el de bayes si la info no es 50-50, o en caso que si como arreglarlo, no se hará nada al respecto.

6. Separar las variables

```python
xd = bdatos[["Business", "Age", "Female", "Years", "Certificates", "Feedback", "Salary", "NPS"]]
yd = bdatos["College"]
```

7. Implementación del modelo

```python
arbol = DecisionTreeClassifier(criterion = "gini", max_depth=4)
#se deciden usar 4 niveles debido a la recomendación del profe
#de usar un nivel por variable hasta llegar a un maximo de 4 especificados
#en el requerimiento del entregable.

arbol.fit(xd, yd)
```

```python
#Observarlo graficamente

variables = ["Business", "Age", "Female", "Years",
             "Certificates", "Feedback", "Salary", "NPS"]

dot_data = export_graphviz(arbol, feature_names = variables)

grafica = graph_from_dot_data(dot_data)
grafica.write_png("Arbol de decision ANEG reto S5.png")
```

8. Evaluar modelo

```python
ydp = arbol.predict(xd)
confusion = confusion_matrix(yd, ydp)
print(confusion)

VN = confusion[0,0]

FP = confusion[0,1]

FN = confusion[1,0]

VP = confusion[1,1]

Exactitud = (VP+VN)/(VP+VN+FP+FN)

Sensibilidad = VP/(VP+FN)

Especificidad = VN/(VN+FP)

Precisión = VP/(VP+FP)

print("Exactitud: ", Exactitud.round(3)*100, "%")
print("Sensibilidad: ", Sensibilidad.round(3)*100, "%")
print("Especificidad: ", Especificidad.round(3)*100, "%")
print("Precisión: ", Precisión*100, "%")


Tasa_de_error = (FP+FN)/(VP+VN+FP+FN)

print("Tasa de error: ", Tasa_de_error.round(3)*100, "%")

Prediccion_negativa = VN/(VN+FN)

print("Predicción negativa: ", Prediccion_negativa.round(3)*100, "%")
```

```python
#Nota, asegurarse de correr la parte anterior antes de esta, ya que al final
#de esta se vuelve a convertir de números a nombres el hardware y software

#tomar variables
xd2 = bdatos2[["Business", "Age", "Female", "Years", "Certificates", "Feedback", "Salary", "NPS"]]

#prediccion
pronostico = arbol.predict(xd2)

print(pronostico)

bdatos2["Pronostico"] = pronostico
bdatos2["Business"] = bdatos2["Business"].map({0: "Hardware", 1: "Software"})


print(bdatos2)

bdatos2.to_excel("Pronostico reto 2.xlsx")
```
